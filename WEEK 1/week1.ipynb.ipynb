```python
# Select the desired sheet from the existing workbook
existing_sheet = existing_workbook.active
```

# Question 4
##  Python program for Prepare Yoshops Survey and Order excel charts Like = Pie Chart and Bar Chart Weekly, Monthly and Yearly Reports.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
df = pd.read_excel('Yoshops Survey_1786_Updated_13_March.xlsx') 

# Define function to generate pie chart
def generate_pie_chart(data, title):
    plt.figure(figsize=(8, 6))
    plt.pie(data.values, labels=data.index, autopct='%1.1f%%')
    plt.axis('equal')
    plt.title(title + " (Pie Chart)")
    plt.show()

# Define function to generate bar chart
def generate_bar_chart(data, title):
    plt.figure(figsize=(10, 6))
    data.plot(kind='bar')
    plt.xlabel(title)
    plt.ylabel("Count")
    plt.show()
    
     


# Generate pie chart for Favorite Food Biryani
favorite_food_data = df['7. What is your Favourite food biryani essay?'].value_counts()
generate_pie_chart(favorite_food_data, "Favorite Food Biryani")

# Generate bar chart for Favorite Food Biryani
generate_bar_chart(favorite_food_data, "Favorite Food Biryani")

# Generate pie chart for Financial Products
financial_products_data = df['9. Which Financial products you like most'].value_counts()
generate_pie_chart(financial_products_data, "Financial Products")

# Generate bar chart for Financial Products
generate_bar_chart(financial_products_data, "Financial Products")

# Generate pie chart for Interest in Yoshops Finance Product
interest_data = df['10. Are you interested in Yoshops Finance  Product.'].value_counts()
generate_pie_chart(interest_data, "Interest in Yoshops Finance Product")

# Generate bar chart for Interest in Yoshops Finance Product
generate_bar_chart(interest_data, "Interest in Yoshops Finance Product")
```
